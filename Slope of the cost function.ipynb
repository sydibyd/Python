#### Notebook import and packages

```python
import matplotlib.pyplot as ppl
import numpy as np

%matplotlib inline
```

#### The cost function
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1  
```

```python
# make 100 individual data points between -3 and3
X_1 = np.linspace(start = -3, stop = 3, num = 5000)
```

```python
# plot data
ppl.plot(X_1, f(X_1))
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('f(x)', fontsize = 12)
ppl.show()
```

#### Slope and Derivative
Create a slope function for the derivative of f(x).

```python
def df(x):
    return 2 * x + 1
```

```python
# plot both function and derivative
ppl.figure(figsize = [15,5])

# Cost function
ppl.subplot(1, 2, 1)
ppl.plot(X_1, f(X_1))
ppl.title('Cost function')
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('f(x)', fontsize = 12)

# Derivative
ppl.subplot(1, 2, 2)
ppl.plot(X_1, df(X_1), color = 'red')
ppl.title('Slope of the cost function')
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('df(x)', fontsize = 12)
ppl.grid()

ppl.show()
```

#### Loops and Gradient Descent

```python
# set a simple for loop
for i in range(5):
    print("Loop", i)
    
print('End')
```

```python
# set a simple while loop
w = 0
while w <= 10:
    print('Loop', w)
    w += 1

print('End')
```

```python
# Gradient Descent
fresh_x = 3
old_x = 0
step_mixer = 0.1
target = 0.000099

# graphing by lists
x_list = [fresh_x]
slope_list = [df(fresh_x)]


for i in range (999):
    old_x = fresh_x
    gradient = df(old_x)
    fresh_x = old_x - step_mixer * gradient
    
    # appending for visuals
    x_list.append(fresh_x)
    slope_list.append(df(fresh_x))
    
    # set a target (precision) to stop the loop
    steps = abs(fresh_x - old_x)
    if steps < target:
        print('loop counter on : ', i)
        break
    
    
print('Local minimum occurs at: ', fresh_x)
print('Slope of df(x) is: ', df(fresh_x))
print('f(x) value for cost is: ', f(fresh_x))
```

```python
# Place the gradient descent on plot
ppl.figure(figsize = [20,4])

# Cost function
ppl.subplot(1, 3, 1)
ppl.plot(X_1, f(X_1), alpha = 0.75)
ppl.title('Cost function')
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('f(x)', fontsize = 12)

dots = np.array(x_list)
ppl.scatter(x_list, f(dots), color = 'red', s=50, alpha = 0.75)

# Derivative
ppl.subplot(1, 3, 2)
ppl.title('Slope of the cost function')
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('df(x)', fontsize = 12)
ppl.grid()

ppl.plot(X_1, df(X_1), color = 'blue', alpha = 0.50)

ppl.scatter(x_list, slope_list, color='red', s=50, alpha=0.75)


# Close Up 
ppl.subplot(1, 3, 3)
ppl.title('Close Up on our goal')
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('df(x)', fontsize = 12)
ppl.xlim(-0.55, -0.2)
ppl.ylim(-0.3, 0.8)
ppl.grid()

ppl.plot(X_1, df(X_1), color = 'blue', alpha = 0.50)
ppl.scatter(x_list, slope_list, color='red', s=500, alpha=0.25)


ppl.show()
```
