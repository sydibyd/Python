## Correlation between different features
#### Strenght and Direction of Correlation <br>
We want to identify features that are correlated with the house prices.

$\rho _{XY} = corr(X,Y)$  

$-1.0 \leq \rho _{XY} \leq +1.0 $

```python
# calculate the correlation between the avg number of rooms and the house price
data["PRICE"].corr(data["RM"])
```

```python
# comparing to the PTRATIO (quality of education)
data["PRICE"].corr(data["PTRATIO"])
```

```python
# calculate all the correlations at the same time
data.corr()
```

#### Let's hide half of the table
Using filter from numpy module

```python
filter_mask = np.zeros_like(data.corr())
tringl_layer = np.triu_indices_from(filter_mask)
filter_mask[tringl_layer] = True
filter_mask
```

```python
# plot the filter mask
# using heatmap to compare stronge positive and negative correlations
plt.figure(figsize=(8,8))
sns.heatmap(data.corr())
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(), mask = filter_mask, annot = True)
sns.set_style('dark')
plt.show()
```

## Advanced visualization
#### plot the relationship between PRICE and NOX


```python
price_crime_corr = round(data['PRICE'].corr(data['CRIM']),4)
plt.figure(figsize=(16,9))
plt.title(f'Crime and Housing Prices (correlation {price_crime_corr})', fontsize = 18)
plt.scatter(x=data['PRICE'], y=data['CRIM'], alpha = 0.25, s=128, color='#099090')
plt.xlabel('Housing Prices')
plt.ylabel('Crime')
plt.show()
```

#### Use the seaborn module to add a new layer to our visualization

```python
sns.set()
sns.set_style('darkgrid')
sns.jointplot(x=data['PRICE'], y=data['CRIM'], height=8, alpha=0.25, joint_kws={'alpha':0.25})
```

#### Aggregating the data points with the kind keyword argument

```python
sns.set()
sns.set_style('darkgrid')
sns.jointplot(x=data['PRICE'], y=data['CRIM'], kind = 'hex', height=8)
```
