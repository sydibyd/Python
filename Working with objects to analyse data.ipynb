```python
import pandas as pp
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as ppl


# read the data
data = pp.read_csv('lsd-math-score-data.csv')

# extract as dataframe [[]]
Time  = data [['Time_Delay_in_Minutes']]
LSD   = data [['LSD_ppm']]
Score = data [['Avg_Math_Test_Score']]
```

```python
reg = LinearRegression()
reg.fit(LSD,Score)
```

```python
# look at attributes
reg.coef_
```

```python
# access the values of an array
print('Theta1 :', reg.coef_[0][0]) ## [0] helps with accessing to the value
print('Intercept :', reg.intercept_[0])
print('R-Square :', reg.score(LSD,Score))
```

#### Plot the data

```python
%matplotlib inline
ppl.scatter(LSD, Score, color = 'blue', s = 96, alpha = 0.75)
# add title, x and y label
ppl.title('Aritmethics cs LSD', fontsize = 14)
ppl.xlabel('LSD ppm', fontsize = 10)
ppl.ylabel('Perfomance', fontsize = 10)
ppl.ylim(25, 85)
ppl.xlim(0,8)
ppl.style.use('seaborn')
```

#### Plot our prediction from our Linear Regression ML model

```python
%matplotlib inline
ppl.scatter(LSD, Score, color = 'blue', s = 96, alpha = 0.75)
# add title, x and y label
ppl.title('Aritmethics cs LSD', fontsize = 14)
ppl.xlabel('LSD ppm', fontsize = 10)
ppl.ylabel('Perfomance', fontsize = 10)
ppl.ylim(25, 85)
ppl.xlim(0,8)
ppl.style.use('seaborn')


predicted_score = reg.predict(LSD)
ppl.plot(LSD, predicted_score, color = 'red')
```
