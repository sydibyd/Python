#### Notebook import and packages

```python
import matplotlib.pyplot as ppl
import numpy as np

%matplotlib inline
```

#### The cost function
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1  
```

```python
# make 100 individual data points between -3 and3
X_1 = np.linspace(start = -3, stop = 3, num = 5000)
```

```python
# plot data
ppl.plot(X_1, f(X_1))
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('f(x)', fontsize = 12)
ppl.show()
```

#### Slope and Derivative
Create a slope function for the derivative of f(x).

```python
def df(x):
    return 2 * x + 1
```

```python
# plot both function and derivative
ppl.figure(figsize = [15,5])

# Cost function
ppl.subplot(1, 2, 1)
ppl.plot(X_1, f(X_1))
ppl.title('Cost function')
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('f(x)', fontsize = 12)

# Derivative
ppl.subplot(1, 2, 2)
ppl.plot(X_1, df(X_1), color = 'red')
ppl.title('Slope of the cost function')
ppl.xlabel('X', fontsize = 12)
ppl.ylabel('df(x)', fontsize = 12)
ppl.grid()

ppl.show()
```
