```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as ppl
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('9249290-cost-revenue-clean.csv')
```

```python
data.describe()
```

Creating two more objects - pandas data frame

```python
x = DataFrame (data, columns = ['production_budget_usd'])
y = DataFrame (data, columns = ['worldwide_gross_usd'])
```

Matplotlib - let's visualize them 

```python
ppl.figure(figsize = (16,9))
ppl.scatter(x,y, alpha = 0.25)
ppl.title('Film Cost vs Global Revenue')
ppl.xlabel('Budget $')
ppl.ylabel('Revenue $')
ppl.ylim(0, 3000000000)
ppl.xlim(0, 450000000)
ppl.show()
```

Add the code to run the Linear Regression

```python
regression = LinearRegression()
regression.fit(x,y)
```

Look at the coefficient and intercept

```python
regression.coef_
```

```python
regression.intercept_
```

```python
ppl.figure(figsize = (16,9))
ppl.scatter(x,y, alpha = 0.25)

ppl.plot(x, regression.predict(x), color = 'black', alpha = 0.75, linewidth = 5)

ppl.title('Film Cost vs Global Revenue')
ppl.xlabel('Budget $')
ppl.ylabel('Revenue $')
ppl.ylim(0, 3000000000)
ppl.xlim(0, 450000000)
ppl.show()
```
